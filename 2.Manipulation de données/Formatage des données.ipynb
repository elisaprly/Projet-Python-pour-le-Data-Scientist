### Quatrième étape
Formatage des données prix, durée et escale pour que ce soit des données quantitatives

```python
price = list(df['Prix'])
duration = list(df ['Duree'])
```

```python
def formatage_prix(liste):
    l=[]
    for i in range (len(liste)):
        c=''
        if len(liste[i]) == 6 :
            c = float(liste[i][0]+liste[i][1]+price[i][2])
        elif len(liste[i]) == 5 : 
            c = float(liste[i][0]+liste[i][1])
        elif len(liste[i]) == 8:
            c = float(liste[i][0]+liste[i][2]+liste[i][3]+ liste[i][4])
        else:
            c = 'NaN'
        l.append(c)
    return l
```

```python
prix = formatage_prix(price)
```

```python
def formatage_duree(liste):
    l=[]
    for i in range (len(liste)):
        l.append(float(duration[i].split('h')[0])+float(duration[i].split('h')[1][1:3])/60)
    return l   
```

```python
duree = formatage_duree(duration)
```

```python
del df['Prix']
del df['Duree']
```

```python
df['Prix'] = prix
df['Duree de vol'] = duree
```

```python
def formatage_escale(df):
    a = pd.DataFrame([df["Total d'escales"] != "direct ", df["Total d'escales"] != "O changement "])
    df1 = df.copy()
    df["Total d'escales"] = a.all().astype(np.int64) + (df1["Total d'escales"] == "2 escales ").astype(np.int64) + (df["Total d'escales"] == "3 escales ").astype(np.int64)
    return(df)

df1 = formatage_escale(df)
```
